```python
from numpy import *
import numpy as np
from math import *
import einsteinpy as ep
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def File_To_Array(Name,Array):
    with open(Name, 'r') as f:
        lines = f.readlines()
    for i in lines:
        Strings=i.replace(",",".")
        Array.append([float(Strings.split()[0]),float(Strings.split()[1])])
    f.close()
```

```python
Argonio=[]
File_To_Array("argonio.txt",Argonio)
Fluorescente=[]
File_To_Array("fluorescente(mercurio).txt",Fluorescente)
Hidrogenio1=[]
File_To_Array("Hidrogenio.txt",Hidrogenio1)
Hidrogenio2=[]
File_To_Array("hidrogenio2.txt",Hidrogenio2)
Incandecente=[]
File_To_Array("incandecente.txt",Incandecente)
LED_Azul=[]
File_To_Array("LED azul.txt",LED_Azul)
LED_Verde=[]
File_To_Array("LED verde.txt",LED_Verde)
LED_Vermelho=[]
File_To_Array("LED vermelho.txt",LED_Vermelho)
LED_Branco=[]
File_To_Array("LED branco (phosphorus).txt",LED_Branco)
Mercurio=[]
File_To_Array("Mercurio.txt",Mercurio)
Results=[Argonio,Fluorescente,Hidrogenio1,Hidrogenio2,Incandecente,LED_Azul,LED_Verde,LED_Vermelho,LED_Branco,Mercurio]
```

```python
def Find_Maximum(Array):
    N=0
    for i in range(0,len(Array)):
        if Array[i][1]>Array[N][1]:
            N=i
    return N,Array[N][0]
```

```python
Find_Maximum(Argonio)
```

```python
for i in Results:
    print(Find_Maximum(i))
```

```python
def Print_Curve(Array,cor=["b-"],title="Titulo"):
    fig=plt.figure()
    axs=fig.add_subplot()
    j=0
    for R in Array:
        X=[i[0] for i in R]
        Y=[i[1] for i in R]
        axs.plot(X,Y,cor[j])
        axs.set_ylabel("Intensidade")
        axs.set_xlabel("Comprimento de Onda em nm")
        j=j+1
    axs.set_title(title)
```

```python
def Print_Top_5(Sequence,i=5):
    A=Find_Maximum(Sequence)
    print(A)
    if i!=1:
        Print_Top_5(Sequence[:A[0]-20]+Sequence[A[0]+20:],i-1)
```
