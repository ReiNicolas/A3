```python
E=[]
for i in range(1,8):
    E.append(-m_e*e**4/(8*h**2*e0**2)*1/i**2)
E=np.array(E)
E=E/e ##Transformar em eletron volt.
E
```

```python
def Up_Side_Down_Arrow(x,y,l,w):
    X=[]
    Y=[]
    for i in range(0,10**4):
        X.append(x)
        Y.append(y-i*l*10**-4)
    for i in range(0,10**3*5):
        X.append(x-i*w*10**-4)
        Y.append(y-l+i*w*10**-4)
    for i in range(0,10**3*5):
        X.append(x+i*w*10**-4)
        Y.append(y-l+i*w*10**-4)
    return X,Y
```
